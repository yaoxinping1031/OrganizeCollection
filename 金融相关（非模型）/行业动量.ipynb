## 1、验证A 股市场的行业动量 (零投资组合) 

```python
import pandas as pd

df = pd.read_csv("行业分类.csv")
```

```python
df
```

```python
# 将'type'列的字符串转换为列表
df['type'] = df['type'].apply(eval)
```

```python
# 将列表展开，并使用get_dummies进行独热编码
one_hot_encoded = df['type'].apply(lambda x: pd.Series([1] * len(x), index=x)).fillna(0)
```

```python
one_hot_encoded
```

```python
# 将独热编码的结果连接到原始DataFrame
df = pd.concat([df, one_hot_encoded], axis=1)
df
```

```python
df = df.iloc[:, 2:]

df
```

```python
import numpy as np

# 在文本的向量表示上，与 Fasttext 模型相同，我们将一个文本中所有词的词向量的平均值作为该文本的特征向量计算每个文本的特征向量
# 计算每个文本的特征向量
feature_vectors = df.apply(lambda row: np.mean(np.where(row == 1)[0]), axis=1)
```
